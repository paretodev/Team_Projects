# <1>. Deal with Null Values

### 1. fill 'monthly income' with median - because mean is not the centrality value that well represent the most of the member. I used median to fill nan valuse.

```python
from collections import Counter
import numpy as np
from matplotlib import pyplot as plt

data['MonthlyIncome'].fillna(data['MonthlyIncome'].median(),inplace=True)
```

```python
# Long tail distribution - which extends to over 300k though the plot only shows till 40k.
monthly_income = np.array(data['MonthlyIncome'])
monthly_income = monthly_income // 1000

income_hist = Counter(monthly_income)
x_list = [x + 0.5 for x in income_hist.keys()]
y_list = income_hist.values()

plt.figure(figsize=(30,13))
plt.bar(x_list, y_list, 1, edgecolor = ( 0 , 0 , 0 ))
plt.axis([0,50,0,40000])
plt.xticks([ i for i in range(50)])
plt.show()
```

### 2. fill 'number of dependents' with mode value of most occuring value.

```python
# fill in number of dependents with mode of most occuring value
data['NumberOfDependents'].fillna(data['NumberOfDependents'].mode()[0], inplace=True)
```

```python
# No null now !!
print(data.isnull().sum())
```

# <3>. normalize particulararly wide-ranged values.
## But the range of the age is to wide compared to other values that I am going to use.<br>So I will normalize the age values by representing them into z scores.

#### ex). 'NumberOfTime30-59DaysPastDueNotWorse', 'NumberOfTimes90DaysLate', 'NumberOfTime60-89DaysPastDueNotWorse', 'NumberOfDependents'.

```python
mean = np.mean(data['age'])
std = np.std(data['age'])

# copy copy = data['age']
copy = data['age']

# make age column z-score normalized.
data['age'] = np.round((copy - mean)/std,3)
data['age']
```

# <4>. Train the model with 5 picked column and given label

```python
attributes=['age','NumberOfTime30-59DaysPastDueNotWorse','NumberOfTimes90DaysLate','NumberOfTime60-89DaysPastDueNotWorse', 'NumberOfDependents']
sol=['SeriousDlqin2yrs']
X=data[attributes]
y=data[sol]
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from xgboost import XGBClassifier
# fit model no training data
model = XGBClassifier()
model.fit(X,y.values.ravel())
y_pred = model.predict(X)
print("The Accuracy score is : ",accuracy_score(y,y_pred)*100,"%")
print(confusion_matrix(y,y_pred))
```

```python
from sklearn.model_selection import KFold,cross_val_score
kf = KFold(n_splits=5, random_state=None) 
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index] 
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
print(np.mean(cross_val_score(model, X, y.values.ravel(), cv=10))*100)
```
