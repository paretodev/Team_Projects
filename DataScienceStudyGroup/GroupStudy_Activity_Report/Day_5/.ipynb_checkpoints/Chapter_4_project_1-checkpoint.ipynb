### Use Elo ranking to rank all the teams in a sports. How accurately do these Elo ranking predict the result of future contest ?

* ELO ecosystem keeps the total scores within it.
* formula is r'(A) = r(A) + k(S(A) - $\mu$(A))
* if input is given as `{'juventus':2, 'real madrid':3}` the system will judge the winner and change the ranking score
* and at last it will print the table.

```python
rank_scores = {'barcelona' : 100, 'real madrid' : 50, 
               'atletico madrid' : 82, 'liverpool' : 70, 
'valencia' : 60, 'manchester city' : 91}
```

```python
# ex) this is the result represented in dictionary  
match_results_1 = [
{'barcelona': 1, 'real madrid': 3},
{'atletico madrid': 1, 'liverpool':0},
{'manchester city': 0, 'valencia': 2}
    ]

#match_day_2 basically you can feed on and on
```

```python
import math
def sigmoid(x):
  return 1 / (1 + math.exp(-x))

# swing size 'k'
k = 2

for result_dict in match_results_1:
    
    tup_list = [(team, score) for team, score in result_dict.items()]
    
    sorted_list = sorted(tup_list, key= lambda tup: tup[1], reverse = True)
    
    winner = sorted_list[0][0]
    loser = sorted_list[1][0]
    
    # formula is r'(A) = r(A) + k(S(A) - Mu(A))
    
    # winner update
    S_a = 1
    P_a = sigmoid(rank_scores[winner] - rank_scores[loser])
    Mu_a = 1*P_a + (-1)*(1 - P_a)
    
    # loser update
    S_b = -1
    P_b = sigmoid(rank_scores[loser] - rank_scores[winner])
    Mu_b = 1*P_b + (-1)*(1 - P_b)
    
    #     update dict
    rank_scores[winner] = rank_scores[winner] + k*(S_a - Mu_a)
    
    #     update dict
    rank_scores[loser] = rank_scores[loser] + k*(S_b - Mu_b)
    

# show update rank scores
sorted_rank = sorted([(x, round(y)) for x,y in rank_scores.items()], key= lambda tup: tup[1], reverse = True)

import pandas as pd
pd.DataFrame(sorted_rank, 
                   columns=['team','rank score'],
                   index = [i + 1 for i in range(len(sorted_rank))])
```
