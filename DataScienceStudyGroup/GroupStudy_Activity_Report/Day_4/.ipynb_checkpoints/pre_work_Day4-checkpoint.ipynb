### 3-8. [3] What types of outliers might you expect to occur in the following data sets:<br>(a) Student grades.<br>(b) Salary data.<br>(c) Lifespans (for each animals) in Wikipedia.

```python
# a). 만점자, 신청하고 안본 사람, 시험보다가 언어만 보고 나간사람
# b). CEO, 이사 / 임시 파트너쉽으로 페이 받은 사람, 인턴 /
# c). 대부분의 동물 15 ~ 25 주변
```

```python
life_spans = {'giant tortoise':150, 
 'box turtle': 123, 'human': 80, 'pearl mussel': 69, 
 'parrot': 66, 'catfish':61, 'alligator':57,
'eel': 53, 'asian elephant': 49, 'carp': 47, 'snapping turtle': 44, 'camel': 43, 'rhinoceros': 41, 'hippo': 39,
'halibut': 37, 'bear': 37, 'crane': 36, 'seal': 34, 'whale': 33, 'pelican': 32, 'baboon': 31, 'giraffe': 29,
'lion': 28, 'horse': 26, 'anaconda': 24, 'gull': 23, 'great horned owl': 26, 'moose': 24, 'eagle': 24,
'cotton-mouth': 23, 'domestic pigeon': 22.5, 'goose': 22, 'heron':21,
'kangeroo': 19, 'cattle': 18, 'canary': 17, 'cat': 17, 'dog': 16, 'queen ant': 16, 'cicada': 16,
'pike': 14, 'wolf': 14, 'sheep': 13, 'reindeer':12, 'cod':11, 'blackbird': 11, 'squirrel': 9, 'fox': 8,
 'starling': 7, 'wood-pecker': 7, 'worker ant': 6, 'hare': 6, 'rat': 4, 'mole':3, 'shrew':2, 'dragon fly': 1}

# get median, mean
import numpy as np
median = np.median(np.array(list(life_spans.values())))
mean = np.mean(np.array(list(life_spans.values())))
```

```python
from collections import Counter
ls_counter = Counter([int(ls//5) for ls in life_spans.values()])
tuples = ls_counter.items()
zones, freq = zip(*tuples)
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(14, 5)) # figure size
barlist = plt.bar([5*i + 2.5 for i in zones], [k for k in freq], 5, edgecolor= [1,1,1])
barlist[2].set_color('r')
plt.axis([-5, 160, 0, 10])
plt.xticks([5*i for i in range(32)])
plt.xlabel('Life Span')
plt.ylabel('The Number of Animal Species per Interval')
plt.title('Are we outliers in life span amongst animal species?')

# plot median as line
y = np.linspace(0,7.9,100)
x = 0*y+median
plt.plot(x, y, '-r', label='median')

# plot mean as line
y = np.linspace(0,3.9,100)
x = 0*y+mean
plt.plot(x, y, '-k', label='mean')

# annotate
plt.annotate('We are here!', xy = (82,1.2), xytext = (-27, 0), textcoords = 'offset points')

# annotate turtles with -> pointed 
plt.annotate('Box Turtle', xy = (118,1.2), xytext = (0, 0), textcoords = 'offset points')
plt.annotate('  Giant\nTortoise', xy = (148,1.2), xytext = (0, 0), textcoords = 'offset points')
plt.annotate(f'median = {median}', xy = (18,8.5), xytext = (-5, 0), textcoords = 'offset points')
plt.annotate(f'mean = {round(mean,1)}', xy = (27.5,4), xytext = (-16, 8), textcoords = 'offset points')
plt.show() # 악어, 메기, 앵무새, 진주홍합 => 상자거북
```

### 3-15. [5] In general, how would you screen for outliers, and what should you do if you find one?

#### We can visualize the data by using graphs and find candidates for outliers.<br><br>1. We can analyze the data and can practically find minimum and maximum values of the points in various dimensions.<br>2. For Normal distributions if the data is a number of sigma away from mean, then that point can be an outlier.<br>3. We can cluster the data points and if there is a point which is very far from cluster center, then it can be an outlier.


#### Handling of outliers: <br>If the outliers are present because of some measurement error or any other error in data collection, then the outliers can be deleted which can lead to model improvement.If the outliers present are not because of some error, then they should not be deleted as when the outlier will come again in test data then our model will give worse results if we have deleted them in training data.
